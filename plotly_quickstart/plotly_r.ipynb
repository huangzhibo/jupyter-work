## R ggplot2导出为plotly json

```r
library(ggplot2)
library(plotly)
```

###  画一个 ggplot2 图

```r
p_gg <- ggplot(mtcars, aes(x = wt, y = mpg, color = factor(cyl))) +
  geom_point(size = 3) +
  theme_minimal()
```

```r
p_gg
```

### 转成 plotly 对象

```r
p_plotly <- ggplotly(p_gg)
```

```r
p_plotly
```

### 生成plotly json

```r
json_str <- plotly_json(
  p_plotly,
  pretty   = TRUE,
  jsonedit = FALSE
)
```

```r
# 写文件
writeLines(json_str, "ggplot2_plotly.json")
```
