### 基础示例 - 散点图

````python
import plotly.graph_objects as go
import numpy as np
````

````python
# 生成示例数据
x = np.random.randn(50)
y = np.random.randn(50)

# 创建图表
fig = go.Figure()
fig.add_trace(go.Scatter(
    x=x,
    y=y,
    mode='markers',
    name='Sample Data',
    marker=dict(size=10, color='blue')
))
    
# 设置标题和轴标签
fig.update_layout(
    title='My First Plotly Chart',
    xaxis_title='X Values',
    yaxis_title='Y Values'
)
````

### WebGL加速散点图

````python
x = np.random.randn(100000)
y = np.random.randn(100000)

# 使用 Scattergl 渲染
fig = go.Figure(data=go.Scattergl(
    x=x, 
    y=y, 
    mode='markers',
    marker=dict(size=5, color='blue')
))
fig.update_layout(title='WebGL 加速散点图')
# fig.show()
````

### 导出为JSON

````python
json_str = fig.to_json()
with open('my_plot2.json', 'w') as f:
    f.write(json_str)
````

### 保存为静态图片 (需要安装kaleido: pip install kaleido)

````python
fig.write_image("my_plot.png")
fig.write_image("my_plot.pdf")
````

## 核心特性
- **交互性**: 自动支持缩放、平移、悬停等交互功能
- **响应式**: 图表自动适应容器大小
- **多格式输出**: 支持HTML、PNG、PDF、SVG等格式
- **Web就绪**: 生成的图表可直接嵌入网页
